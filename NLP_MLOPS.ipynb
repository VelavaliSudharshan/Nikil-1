```python
from google_play_scraper import Sort, reviews_all

result_FYER = reviews_all(
    'com.fyers.trader',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='IN', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    filter_score_with=5 # defaults to None(means all score)
)
```

```python

result_UPSTOX =reviews_all(
    'in.upstox.app',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='IN', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    filter_score_with=5 # defaults to None(means all score)
)
```

```python
result =result_FYER +result_UPSTOX
```

```python
spark = SparkSession.builder.master("local[*]").appName("AppendDataToDataFrame").getOrCreate()

# Define the schema for the DataFrame
schema = StructType([
    StructField("reviewId", StringType(), True),
    StructField("userName", StringType(), True),
    StructField("userImage", StringType(), True),
    StructField("content", StringType(), True),
    StructField("score", IntegerType(), True)
])
```
